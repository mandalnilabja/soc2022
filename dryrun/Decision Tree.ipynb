```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv'
data = pd.read_csv(url)
```

```python
X = data.drop(['Survived', 'PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
X
```

```python
y = data.Survived
y
```

```python
X.Age.isnull().values.any()
```

```python
X.Age = X.Age.fillna(X.Age.median())
```

```python
from sklearn.preprocessing import LabelEncoder
X['Sex'] = LabelEncoder().fit_transform(X.Sex)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier().fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score
display(model.score(X_test, y_test))

y_pred = model.predict(X_test)
display(f1_score(y_test, y_pred))
```
