```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
wine=load_wine()
```

```python
dir(wine)
```

```python
from sklearn.preprocessing import MinMaxScaler
df=MinMaxScaler().fit_transform(wine.data)
```

```python
df=pd.DataFrame(data=df, columns=wine.feature_names)
df['target']=wine.target
df.isnull().values.any()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'], test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB().fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model_Mn=MultinomialNB().fit(X_train,y_train)
```

```python
model_Mn.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
CV_G=cross_val_score(GaussianNB(), X_train, y_train, cv=3)
CV_Mn=cross_val_score(MultinomialNB(), X_train, y_train, cv=3)

display(CV_G, CV_Mn)
```

Multinomial is Better than Gaussian
