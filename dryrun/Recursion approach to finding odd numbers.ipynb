```python
class Solution(object):
    def countOdds(self, low, high):
        """
        :type low: int
        :type high: int
        :rtype: int
        """
        count=0
        if high==low and low%2==1:
            count=count+1
        else:
            if low%2==1:
                count=count+1
            if high%2==1:
                count=count+1
            count=count+Solution.oddnum(self, low, high)
        return count        
        
        
    def oddnum(self, low, high):        
        count=0
        if (high-low)<=1:
            return 0
        else:
            mid=int((high+low)/2)
            if mid%2==1:
                count=count+1
            count=count+Solution.oddnum(self,mid,high)
            count=count+Solution.oddnum(self,low,mid)
            return count
```
